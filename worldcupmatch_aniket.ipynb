```python
# Assuming 'data' is your DataFrame
team1 = data['Away Team Name']

# Count the occurrences of each team
team_counts = team1.value_counts()

# Plot the team counts as a bar chart
plt.bar(team_counts.index, team_counts.values)
plt.xlabel('Team')
plt.ylabel('Number of Occurrences')
plt.title('Frequency of Away Team Name') # Changed plt.team to plt.title
plt.xticks(rotation=90)  # Rotate x-axis labels for readability if needed
plt.tight_layout()  # Adjust layout to prevent labels from overlapping
plt.show()
```

```python
sns.barplot(x='City', y='Home Team Name', data=data) # Changed 'HomeTownName' to 'Home Team Name'
plt.xlabel('City')
plt.ylabel('Home Team Name') # Changed 'Home Town Name' to 'Home Team Name'
plt.title('City or Home Town name')
plt.xticks(rotation=45)
plt.show()
```
